# Splitting our data into training and testing

```python
df=df[["symboling","wheelbase","carlength","carwidth","carheight","curbweight","enginesize","boreratio","stroke","compressionratio","horsepower","peakrpm","citympg","highwaympg","price"]]
```

```python
x=np.array(df.drop('price',axis=1))
y=np.array(df['price'])
```

```python
from sklearn.model_selection import train_test_split
from sklearn.preprocessing import MinMaxScaler
from sklearn.tree import DecisionTreeRegressor
from sklearn import metrics
```

```python
x_train,x_test,y_train,y_test=train_test_split(x,y,test_size=0.2)
```

# Performing scaling on data using MinMax Scaler

```python
scaler=MinMaxScaler()
x_train=scaler.fit_transform(x_train)
x_test=scaler.transform(x_test)
```

# Training the model

```python
model=DecisionTreeRegressor()
model.fit(x_train,y_train)
predictions=model.predict(x_test)
```

# Evaluating the model

```python
print(f"Accuracy            : {model.score(x_test,predictions)*100}")
print(f"R Squared Error     :{round(metrics.r2_score(y_test,predictions)*100,2)}")
print(f"Mean absolute Error :{metrics.mean_absolute_error(y_test,predictions)}")
print(f"Mean Squared Error  :{metrics.mean_squared_error(y_test,predictions)}")
print(f"Root Mean Square Error :{np.sqrt(metrics.mean_squared_error(y_test,predictions))}")
```

```python
# Plotting actual and Predicted price
```

```python
plt.scatter(y_test,predictions)
plt.xlabel('Actual Price')
plt.ylabel('Predicted Price')
plt.title('Actual price vs Predicted price')
plt.show()
```

```python
#Testing the model with sample data
def prediction(input_data):
    #Input data
    car_data=df.drop('price',axis=1).iloc[input_data]
    
    #Scaling Input Data
    car_scaler=scaler.transform(car_data.values.reshape(-1,14))
    
    #Predict the input data
    prediction=model.predict(car_scaler)
    
    #Predicted price of input data
    print(f"Predicted price of the car is : {prediction[0]}")
    
    print()
    
    #Actual price of input data 
    real_price=df.iloc[input_data].price
    print(f"Real price of the car is       :{real_price}")
    
prediction(12)
```
